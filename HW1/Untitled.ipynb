```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_parquet("green_tripdata_2025-11.parquet")
```

```python
df.describe()
```

```python
df["DT_lpep_pickup_datetime"] = pd.to_datetime(df["lpep_pickup_datetime"], errors="coerce")
```

```python
df["DT_lpep_pickup_datetime"].isna().sum()
```

```python
A = pd.to_datetime("2025-11-01")
B = pd.to_datetime("2025-12-01")

mask = (df["DT_lpep_pickup_datetime"] >= A) & \
    (df["DT_lpep_pickup_datetime"] < B) & \
    (df["trip_distance"] <= 1)
```

```python
mask.sum()
```

```python
df_temp = df.loc[df["trip_distance"]<100,["trip_distance","DT_lpep_pickup_datetime"]]
df_temp.sort_values("trip_distance",ascending=False)
```

```python
df_temp = df.groupby("PULocationID", as_index=False)["total_amount"].sum()
```

```python
df_temp.sort_values("total_amount",ascending=False)
```

```python
df_temp = df.loc[(df["DT_lpep_pickup_datetime"] >= A) & \
    (df["DT_lpep_pickup_datetime"] < B) & \
    (df["PULocationID"]==74),
    ["tip_amount","DOLocationID"]]
df_temp.sort_values("tip_amount",ascending=False)
```
